```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('CC GENERAL.csv')
data
```

```python
data.isnull().sum()
```

```python
data['CREDIT_LIMIT'].value_counts()
```

```python
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean() , inplace = True)
```

```python
data['MINIMUM_PAYMENTS'].value_counts()
```

```python
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean() , inplace = True)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
data=data.drop("CUST_ID", axis=1)
data
```

```python
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')

clust_labels=model.fit_predict(data)
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt

fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data['CASH_ADVANCE_TRX'] , data["CASH_ADVANCE"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("CASHADVANCETRX")
ax.set_ylabel("CASH_ADVANCE_TRX")
plt.colorbar(scatter)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10,7))
plt.title("Dendrograms")
dend=shc.dendrogram(shc.linkage(data.head(50) , method="complete"))
```
